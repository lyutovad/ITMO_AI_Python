## Часть 1
Преобразование строки в число с обработкой ошибок

```python
def parse_int(s: str) -> int | None:
    try:
        return int(s.strip())
    except ValueError:
        return None


examples = ["12", " -7 ", "12a", ""]
for value in examples:
    print(parse_int(value))
```

## Часть 2
Функции для работы со списками чисел

```python
nums = [12, -5, 0, 8, 17, -21, 34, 6, -2, 15, 9, 10, -14, 27, 3, 18, -7, 24, 1, -9, 30, 11]


def count_positive(nums: list[int]) -> int:
    count = 0
    for num in nums:
        if num > 0:
            count += 1
    return count


def sum_even(nums: list[int]) -> int:
    total = 0
    for num in nums:
        if num % 2 == 0:
            total += num
    return total


def first_divisible(nums: list[int], k: int) -> int | None:
    result = None
    for num in nums:
        if num % k == 0:
            result = num
            break
    return result


print(count_positive(nums))
print(sum_even(nums))
print(first_divisible(nums, 7))
print(first_divisible(nums, 13))
```

## Часть 3
Сортировка словарей через lambda

```python
items = [
    {"name": "A", "price": 120, "rating": 4.7},
    {"name": "B", "price": 90, "rating": 4.2},
    {"name": "C", "price": 150, "rating": 4.9},
]

sorted_by_price = sorted(items, key=lambda item: item["price"])
sorted_by_rating = sorted(items, key=lambda item: item["rating"], reverse=True)

print(sorted_by_price)
print(sorted_by_rating)
```

## Часть 4
Класс с проверкой цены и методом скидки

```python
class Chocolate:
    def __init__(self, name: str, price: float):
        if price <= 0:
            raise ValueError("price must be greater than 0")
        self.name = name
        self.price = price

    def discount(self, percent: int) -> float:
        return self.price * (1 - percent / 100)


chocolate1 = Chocolate("Milka", 180.0)
chocolate2 = Chocolate("Lindt", 450.0)

print(chocolate1.discount(10))
print(chocolate2.discount(25))
```
