# ЧАСТЬ 1. Работа со строками и исключениями (4 балла)

Создайте функцию:

```python
def parse_int(s: str) -> int | None:
    ...
```


````python
def parse_int(s: str) -> int | None:
    try:
        return int(s.strip()) #Удаляем пробелы и пытаемся конвертировать
    except ValueError:
              return None   #Если возникла ошибка, возвращаем None

test_cases = ["12", " -7 ", "12a", ""]

for test in test_cases:
    result = parse_int(test)
    print(f"Вход: '{test}' -> Результат: {result}")
````

# ЧАСТЬ 2. Циклы и работа со списками (6 баллов)

Создайте список чисел (20–30 элементов).

Реализуйте функции:

```python
def count_positive(nums: list[int]) -> int:
    ...

def sum_even(nums: list[int]) -> int:
    ...

def first_divisible(nums: list[int], k: int) -> int | None:
    ...
```

````python
import random
#Создаем список согласно 
numbers = [random.randint(-50, 50) for _ in range(25)]
print(f"список: {numbers}")
````

````python
#считаем положительные числа
def count_pos(nums: list[int]) -> int:
    count = 0
    for n in nums:
        if n > 0:
            count += 1
    return count

#считаем сумму чётных чисел
def sum_even(nums: list[int]) -> int:
    total = 0
    for n in nums:
        #число чётное, если остаток от деления на 2 равен 0
        if n % 2 == 0:
            total += n
    return total

#находим первое число, делящееся на k
def first_div(nums: list[int], k: int) -> int | None:
    for n in nums:
        if n % k == 0:
            #используем return, который по логике заменяет break и сразу возвращает результат
            return n
    #если цикл закончился и ничего не найдено
    return None

print(f"количество положительных чисел: {count_pos(numbers)}")
print(f"сумма чётных чисел: {sum_even(numbers)}")

#проверим кратность, например, числу 3
k = 3
result_div = first_div(numbers, k)
print(f"первое число, делящееся на {k}: {result_div}")
````

## Требования

### 1. count_positive  
Возвращает количество чисел больше 0.

### 2. sum_even  
Возвращает сумму всех чётных чисел.

### 3. first_divisible  
Возвращает первое число, которое делится на `k`.  
Если таких чисел нет — вернуть `None`.

Использовать:
- цикл `for`
- оператор `break`

### Оценивание

- `first_divisible` (с `break` и корректным `None`) — **2 балла**

# ЧАСТЬ 3. Lambda и сортировки (2 балла)

Дан список словарей:

```python
items = [
  {"name": "A", "price": 120, "rating": 4.7},
  {"name": "B", "price": 90, "rating": 4.2},
  {"name": "C", "price": 150, "rating": 4.9},
]
```

Необходимо:

1. Отсортировать список по `price` (по возрастанию).
2. Отсортировать список по `rating` (по убыванию).

Использовать `sorted(..., key=lambda ...)`.


````python
items = [
  {"name": "A", "price": 120, "rating": 4.7},
  {"name": "B", "price": 90, "rating": 4.2},
  {"name": "C", "price": 150, "rating": 4.9},
]
````

````python
#сортировка по возрастанию
#lambda берет словарь x и возвращает значение по ключу 'price'
items_by_price = sorted(items, key=lambda x: x['price'])

#сортировка по убыванию
#используем reverse=True, чтобы от большего к меньшему
items_by_rating = sorted(items, key=lambda x: x['rating'], reverse=True)
````

````python
print("сортировка по цене -возрастание:")
for item in items_by_price:
    print(item)

print("\nсортировка по рейтингу -убывание:")
for item in items_by_rating:
    print(item)
````


### Оценивание

- корректная сортировка по `price` — **2 балл**
- корректная сортировка по `rating` — **2 балл**

# ЧАСТЬ 4. Мини-ООП (2 балла)

Создайте класс:

```python
class Chocolate:
    ...
```

## Требования

1. Поля:
   - `name: str`
   - `price: float`
2. Если `price <= 0` — выбросить `ValueError`
3. Метод:

```python
def discount(self, percent: int) -> float:
    ...
```

````python
class Chocolate:
    def __init__(self, name: str, price: float):
        if price <= 0:  #проверка если цена<=0,то выбрасываем исключение
            raise ValueError("Цена должна быть больше нуля")
        self.name = name
        self.price = price
    def discount(self, percent: int) -> float:
        return self.price * (1 - percent / 100) #метод возвращает новую цену со скидкой
````

````python
#создание объектов
try:
    choco1 = Chocolate("Alpen Gold", 100)
    choco2 = Chocolate("Milka", 150)   
    # демонстрация метода - discount
    print(f"Шоколад: {choco1.name}, Цена: {choco1.price}, Со скидкой 20%: {choco1.discount(20)}")
    print(f"Шоколад: {choco2.name}, Цена: {choco2.price}, Со скидкой 10%: {choco2.discount(10)}")
    #проверка валидации-вызовет ошибку
    print("\nпроверка некорректной цены:")
    bad_choco = Chocolate("Горький", -5)
except ValueError as e:
    print(f"ошибка: {e}")
````

Метод должен возвращать новую цену со скидкой (не обязательно менять поле).
