## 6. Exercise 5: Functions

Определим функции для задания: `parse_int`, `count_positive`, `sum_even`, `first_divisible` и продемонстрируем.

```python
from typing import List, Optional

# parse_int implementation
def parse_int(s: str) -> Optional[int]:
    s = s.strip()
    try:
        return int(s)
    except ValueError:
        return None

# functions on list

def count_positive(nums: List[int]) -> int:
    count = 0
    for n in nums:
        if n > 0:
            count += 1
    return count

def sum_even(nums: List[int]) -> int:
    total = 0
    for n in nums:
        if n % 2 == 0:
            total += n
    return total

def first_divisible(nums: List[int], k: int) -> Optional[int]:
    for n in nums:
        if n % k == 0:
            return n
    return None

# демонстрация

display(parse_int("12"))
display(parse_int(" -7 "))
display(parse_int("12a"))
display(parse_int(""))

nums = [ -5, 0, 3, 4, 8, -1, 2, 7, 10, -3, 6, 12, 15, 18, -2, 1, 9, 14, -6, 11 ]
display("count_positive", count_positive(nums))
display("sum_even", sum_even(nums))
display("first_divisible by 5", first_divisible(nums, 5))
display("first_divisible by 20", first_divisible(nums, 20))
```

## 9. Mini-OOP: класс Chocolate

Создадим класс с полями `name` и `price`, проверкой цен и методом `discount`.

```python
class Chocolate:
    def __init__(self, name: str, price: float):
        self.name = name
        if price <= 0:
            raise ValueError("Price must be positive")
        self.price = price
    
    def discount(self, percent: int) -> float:
        return self.price * (1 - percent / 100)

# демонстрация
ch1 = Chocolate("Milk", 100)
ch2 = Chocolate("Dark", 150)
display(ch1.name, ch1.price, "discount10", ch1.discount(10))
display(ch2.name, ch2.price, "discount20", ch2.discount(20))
```
