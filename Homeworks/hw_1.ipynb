# ЧАСТЬ 1. Работа со строками и исключениями (4 балла)

Создайте функцию:

```python
def parse_int(s: str) -> int | None:
    ...
```

## Требования:

1. Функция должна пытаться преобразовать строку `s` в `int`.
2. Перед преобразованием необходимо удалить лишние пробелы (`strip()`).
3. Если преобразование невозможно — функция не должна падать, а должна возвращать `None`.
4. Использовать `try/except ValueError`.

## Продемонстрировать работу функции:

- `"12"`
- `" -7 "`
- `"12a"`
- `""`

### Оценивание

- корректное использование `try/except` — **2 балла**
- корректная обработка пробелов — **2 балл**
- демонстрация примеров работы — **2 балл**

---


````python
def parse_int(s: str) -> int | None:
  try:
    return int(s.strip())
  except ValueError:
    return None
````

````python
for s in ["12", " -7 ", "12a", ""]:
  print("Результат работы программы для ", s, ":", parse_int(s))
````

# ЧАСТЬ 2. Циклы и работа со списками (6 баллов)

Создайте список чисел (20–30 элементов).

Реализуйте функции:

```python
def count_positive(nums: list[int]) -> int:
    ...

def sum_even(nums: list[int]) -> int:
    ...

def first_divisible(nums: list[int], k: int) -> int | None:
    ...
```

## Требования

### 1. count_positive  
Возвращает количество чисел больше 0.

### 2. sum_even  
Возвращает сумму всех чётных чисел.

### 3. first_divisible  
Возвращает первое число, которое делится на `k`.  
Если таких чисел нет — вернуть `None`.

Использовать:
- цикл `for`
- оператор `break`

### Оценивание

- `count_positive` — **2 балла**
- `sum_even` — **2 балла**
- `first_divisible` (с `break` и корректным `None`) — **2 балла**

---

````python
import numpy as np

my_list = list(np.random.randint(1, 101, size=20))
````

````python
def count_positive(nums: list[int]) -> int:
  k = 0
  for i in nums:
    if i > 0:
      k+=1
  return k
````

````python
def sum_even(nums: list[int]) -> int:
  k = 0
  for i in nums:
    if i %2 == 0:
      k+=i
  return k
````

````python
def first_divisible(nums: list[int], k: int) -> int | None:
  if k == 0:
    return None
  for i in nums:
    if i %k == 0:
      return i
  return None
````

````python
print("Мой список:", ', '.join([str(i) for i in my_list]))

print("Результат count_positive:", count_positive(my_list))
print("Результат sum_even:", sum_even(my_list))
for i in [1, -9, 100, 5]:
  print("Результат first_divisible для k = ",i, ":", first_divisible(my_list,i))
````

# ЧАСТЬ 3. Lambda и сортировки (2 балла)

Дан список словарей:

```python
items = [
  {"name": "A", "price": 120, "rating": 4.7},
  {"name": "B", "price": 90, "rating": 4.2},
  {"name": "C", "price": 150, "rating": 4.9},
]
```

Необходимо:

1. Отсортировать список по `price` (по возрастанию).
2. Отсортировать список по `rating` (по убыванию).

Использовать `sorted(..., key=lambda ...)`.

### Оценивание

- корректная сортировка по `price` — **2 балл**
- корректная сортировка по `rating` — **2 балл**

---

````python
items = [
  {"name": "A", "price": 120, "rating": 4.7},
  {"name": "B", "price": 90, "rating": 4.2},
  {"name": "C", "price": 150, "rating": 4.9},
]
````

````python
sorted(items, key = lambda point: point["price"])
````

# ЧАСТЬ 4. Мини-ООП (2 балла)

Создайте класс:

```python
class Chocolate:
    ...
```

## Требования

1. Поля:
   - `name: str`
   - `price: float`
2. Если `price <= 0` — выбросить `ValueError`
3. Метод:

```python
def discount(self, percent: int) -> float:
    ...
```

Метод должен возвращать новую цену со скидкой (не обязательно менять поле).

## Продемонстрировать:

- создание минимум двух объектов
- вызов метода `discount`

### Оценивание

- корректная структура класса — **2 балл**
- корректная логика метода и демонстрация — **2 балл**

---

````python
class Chocolate:
  def __init__(self, name, price):
    try:
      self.name = str(name)
    except:
      self.name = None

    try:
      self.price = float(price)
    except:
      raise ValueError("Цена должна быть float")

    try:
      if self.price <= 0:
        raise ValueError("Цена должна быть положительной")
    except ValueError as e:
      print(e)
      self.price = None
  def discount(self, percent: int) -> float | None:
      if self.price is None:
          return None
      if not (0 <= percent <= 100):
          return None
      return self.price * (100 - percent)/100
````

````python
alpen_gold = Chocolate("Alpen Gold", 79)
milka = Chocolate("Milka", 119)
````

````python
alenka = Chocolate("Аленка", -100)
````

````python
alenka = Chocolate("Аленка", '100')
````

````python
for disc in [10, 20, 90]:
  for choko in [alpen_gold, milka]:
    print("Шоколадка ", choko.name, " со скидкой ", disc, "% будет стоить ", choko.discount(disc))
````
