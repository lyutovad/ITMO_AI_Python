ЧАСТЬ 1. Работа со строками и исключениями

```python
def parse_int(s: str) -> int | None:
    try:
        return int(s.strip())
    except ValueError:
        return None

# Демонстрация работы
test_cases = ["12", " -7 ", "12a", ""]

for case in test_cases:
    result = parse_int(case)
    print(f"case: \"{case}\"; result: {result}")
```

ЧАСТЬ 2. Циклы и работа со списками

```python
nums = [-5, 1, 3, 8, -2, 15, 7, 12, -9, 4,
        21, -1, 6, 145, 18, -3, 11, 2, -7, 9,
        14, -4, 1, 16, 9]

def count_positive(nums: list[int]) -> int:
    count = 0
    for n in nums:
        if n > 0:
            count += 1
    return count


def sum_even(nums: list[int]) -> int:
    total = 0
    for n in nums:
        if n % 2 == 0:
            total += n
    return total


def first_divisible(nums: list[int], k: int) -> int | None:
    for n in nums:
        if n % k == 0:
            # break не нужен, т.к. return явно вернет ответ и закончит выполнение функции
            return n
    return None


# Демонстрация
print(f"Количество положительных чисел : {count_positive(nums)}")
print(f"Сумма чётных чисел             : {sum_even(nums)}")
print(f"Первое число, делящееся на 3   : {first_divisible(nums, 3)}")
print(f"Первое число, делящееся на 7   : {first_divisible(nums, 7)}")
print(f"Первое число, делящееся на 100 : {first_divisible(nums, 100)}")
```

ЧАСТЬ 3. Lambda и сортировки (2 балла)

```python
items = [
    {"name": "A", "price": 120, "rating": 4.7},
    {"name": "B", "price": 90,  "rating": 4.2},
    {"name": "C", "price": 150, "rating": 4.9},
]

sorted_by_price  = sorted(items, key=lambda x: x["price"])
sorted_by_rating = sorted(items, key=lambda x: x["rating"], reverse=True)

print("Сортировка по цене (возрастание):")
for item in sorted_by_price:
    print(f"  {item}")

print("\nСортировка по рейтингу (убывание):")
for item in sorted_by_rating:
    print(f"  {item}")
```

ЧАСТЬ 4. Мини-ООП

```python
class Chocolate:
    def __init__(self, name: str, price: float):
        if price <= 0:
            raise ValueError(f"Цена должна быть больше 0, получено: {price}")
        self.name = name
        self.price = price

    def discount(self, percent: int) -> float:
        if not (0 <= percent <= 100):
            raise ValueError("Скидка должна быть от 0 до 100%")
        return self.price * (1 - percent / 100)

    def __repr__(self):
        return f"Chocolate(name={self.name!r}, price={self.price})"


# Демонстрация — корректные объекты
choco1 = Chocolate("Milka", 199.99)
choco2 = Chocolate("Lindt", 350.0)

print(choco1)
print(f"  Цена после скидки 10%: {choco1.discount(10):.2f}")
print(f"  Цена после скидки 25%: {choco1.discount(25):.2f}\n")
print(choco2)
print(f"  Цена после скидки 15%: {choco2.discount(15):.2f}")

# Демонстрация — некорректная цена
print("\nПопытка создать объект с ценой -50:")
try:
    bad_choco = Chocolate("BadChoco", -50)
except ValueError as e:
    print(f"  Поймано исключение: {e}")
```
