## Часть 1. Работа со строками и исключениями

```python
def parse_int(s: str) -> int | None:
    try:
        return int(s.strip())
    except ValueError:
        return None
```

```python
print(parse_int("12"))
print(parse_int(" -7 "))
print(parse_int("12a"))
print(parse_int(""))
```

## Часть 2. Циклы и работа со списками

```python
nums = [3, -3, 8, 0, -5, 13, 7, -3, 14, 2, -8, 6, 11, 9, 5, -6, 15, 1, -4, 10, 13, -7, 5, 16]
```

```python
def count_positive(nums: list[int]) -> int:
    cnt = 0
    for n in nums:
        if n > 0:
            cnt += 1
    return cnt

print("Положительных:", count_positive(nums))
```

```python
def sum_even(nums: list[int]) -> int:
    s = 0
    for n in nums:
        if n % 2 == 0:
            s += n
    return s

print("Сумма четных:", sum_even(nums))
```

```python
def first_divisible(nums: list[int], k: int) -> int | None:
    if k == 0:
        raise ValueError("k не может быть равен 0")

    for n in nums:
        if n % k == 0:
            result = n
            break
    else:
        return None

    return result

print("Первое делящееся на 4:", first_divisible(nums, 4))
print("Первое делящееся на 100:", first_divisible(nums, 100))
```

## Часть 3. Lambda и сортировки

```python
items = [
    {"name": "A", "price": 120, "rating": 4.7},
    {"name": "B", "price": 90, "rating": 4.2},
    {"name": "C", "price": 150, "rating": 4.9},
]
```

```python
by_price = sorted(items, key=lambda x: x["price"])
print("Цена по возрастанию:")
for item in by_price:
    print(item)
```

```python
by_rating = sorted(items, key=lambda x: x["rating"], reverse=True)
print("Рейтинг по убыванию):")
for item in by_rating:
    print(item)
```

## Часть 4. Мини-ООП

```python
class Chocolate:
    def __init__(self, name: str, price: float):
        if price <= 0:
            raise ValueError("Цена должна быть больше 0")
        self.name = name
        self.price = price

    def discount(self, percent: int) -> float:
        return self.price * (1 - percent / 100)
```

```python
c1 = Chocolate("Bay fox", 250)
c2 = Chocolate("Победа", 180)

print(c1.name, "- скидка 10%:", c1.discount(10))
print(c2.name, "- скидка 25%:", c2.discount(25))
```

```python
try:
    c3 = Chocolate("Test", -50)
except ValueError as e:
    print("Ошибка:", e)
```
