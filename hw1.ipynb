## Часть 1. Работа со строками и исключениями

Функция `parse_int` пытается преобразовать строку в `int`.

```python
def parse_int(s: str) -> int | None:
    try:
        return int(s.strip())
    except ValueError:
        return None
```

Демонстрация работы функции:

```python
print(parse_int("12"))    # 12
print(parse_int(" -7 "))  # -7
print(parse_int("12a"))   # None
print(parse_int(""))      # None
```

## Часть 2. Циклы и работа со списками

Список из 25 чисел.

```python
nums = [3, -1, 7, 0, 12, -5, 8, 4, -3, 15, 6, -9, 2, 11, -2, 18, 0, 5, -7, 9, 14, -4, 1, 10, -6]
```

`count_positive` - считает количество чисел больше нуля с помощью цикла `for`.

```python
def count_positive(nums: list[int]) -> int:
    count = 0
    for n in nums:
        if n > 0:
            count += 1
    return count

print(count_positive(nums))
```

`sum_even` - возвращает сумму всех чётных чисел.

```python
def sum_even(nums: list[int]) -> int:
    total = 0
    for n in nums:
        if n % 2 == 0:
            total += n
    return total

print(sum_even(nums))
```

`first_divisible` - возвращает первое число, делящееся на `k`. Использует `break` для остановки цикла. Если таких чисел нет — возвращает `None`.

```python
def first_divisible(nums: list[int], k: int) -> int | None:
    result = None
    for n in nums:
        if n % k == 0:
            result = n
            break
    return result

print(first_divisible(nums, 4))   # первое число, делящееся на 4
print(first_divisible(nums, 13))  # None — таких чисел нет
```

## Часть 3. Lambda и сортировки

Сортировка списка словарей с помощью `sorted` и лямбда-функций.

```python
items = [
    {"name": "A", "price": 120, "rating": 4.7},
    {"name": "B", "price": 90,  "rating": 4.2},
    {"name": "C", "price": 150, "rating": 4.9},
]
```

Сортировка по `price` по возрастанию:

```python
by_price = sorted(items, key=lambda x: x["price"])
print(by_price)
```

Сортировка по `rating` по убыванию:

```python
by_rating = sorted(items, key=lambda x: x["rating"], reverse=True)
print(by_rating)
```

## Часть 4. Мини-ООП

Класс `Chocolate` представляет шоколадку с названием и ценой. При попытке создать объект с ценой ≤ 0 выбрасывается `ValueError`. Метод `discount` возвращает новую цену с учётом скидки.

```python
class Chocolate:
    def __init__(self, name: str, price: float):
        if price <= 0:
            raise ValueError(f"Цена должна быть больше нуля, получено: {price}")
        self.name = name
        self.price = price

    def discount(self, percent: int) -> float:
        return self.price * (1 - percent / 100)
```

Демонстрация работы класса:

```python
choco1 = Chocolate("Milka", 150.0)
choco2 = Chocolate("Alpen Gold", 99.0)

print(f"{choco1.name}: цена со скидкой 20% = {choco1.discount(20)} руб.")
print(f"{choco2.name}: цена со скидкой 10% = {choco2.discount(10)} руб.")

try:
    bad = Chocolate("Ошибка", -50)
except ValueError as e:
    print(f"Ошибка: {e}")
```
