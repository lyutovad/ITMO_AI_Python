## Часть 1. Работа со строками и исключениями
Функция преобразует строку в число.
Если преобразование невозможно — возвращается None.

```python
def parse_int(s: str) -> int | None:
    try:
        s = s.strip()
        return int(s)
    except ValueError:
        return None
```

```python
print(parse_int("12"))
print(parse_int(" -7 "))
print(parse_int("12a"))
print(parse_int(""))
```

## Часть 2. Циклы и работа со списками

```python
nums = [5, -3, 12, 0, 7, -8, 14, 9, -1, 4, 6, 18, -5, 11, 2, -9, 20, 3, -4, 16]
```

```python
def count_positive(nums: list[int]) -> int:
    count = 0
    for n in nums:
        if n > 0:
            count += 1
    return count
```

```python
def sum_even(nums: list[int]) -> int:
    total = 0
    for n in nums:
        if n % 2 == 0:
            total += n
    return total
```

```python
def first_divisible(nums: list[int], k: int) -> int | None:
    result = None

    for n in nums:
        if n % k == 0:
            result = n
            break

    return result
```

```python
print(count_positive(nums))
print(sum_even(nums))
print(first_divisible(nums, 3))
print(first_divisible(nums, 17))
```

## Часть 3. Lambda и сортировки

```python
items = [
  {"name": "A", "price": 120, "rating": 4.7},
  {"name": "B", "price": 90, "rating": 4.2},
  {"name": "C", "price": 150, "rating": 4.9},
]
```

```python
sorted_by_price = sorted(items, key=lambda x: x["price"])
print(sorted_by_price)
```

```python
sorted_by_rating = sorted(items, key=lambda x: x["rating"], reverse=True)
print(sorted_by_rating)
```

## Часть 4. Мини ООП

```python
class Chocolate:

    def __init__(self, name: str, price: float):
        if price <= 0:
            raise ValueError("Price must be positive")

        self.name = name
        self.price = price

    def discount(self, percent: int) -> float:
        return self.price * (1 - percent / 100)
```

```python
ch1 = Chocolate("Milka", 2.5)
ch2 = Chocolate("Lindt", 4.0)

print(ch1.discount(10))
print(ch2.discount(25))
```
