## Самые популярные имена

Из столбца `Name` извлечём собственно имя пассажира и посмотрим самые частые имена среди мужчин и женщин.

```python
import re

def extract_first_name(name):

    # если есть имя в скобках
    match = re.search(r"\((.*?)\)", name)
    if match:
        return match.group(1).split()[0]

    # иначе обычный вариант
    name_part = name.split(",")[1]
    name_part = name_part.split(".")[1]

    return name_part.strip().split()[0]

df["FirstName"] = df["Name"].apply(extract_first_name)

male_name = df[df["Sex"] == "male"]["FirstName"].mode()[0]
female_name = df[df["Sex"] == "female"]["FirstName"].mode()[0]

male_name, female_name
```

## Самые популярные мужские и женские имена в каждом классе

Сгруппируем пассажиров по классу (`Pclass`) и полу (`Sex`),
после чего для каждой группы найдём наиболее частое имя (`mode`).

```python
popular_names = (
    df
    .groupby(["Pclass", "Sex"])["FirstName"]
    .agg(lambda x: x.mode()[0])
)

popular_names
```
