```python
#1 задание
def parse_int(s: str) -> int | None:
    try:
        return int(s.strip())
    except ValueError:
        return None
```

```python
print(parse_int("12"))
print(parse_int(" -7 "))
print(parse_int("12a"))
print(parse_int(""))
```

```python
spisok = [1,65,74,3,4,2,2,97,75,43,52,3,242,42,246,-1,2,-24,9,-170,315]
```

```python
def count_positive(nums: list[int]) -> int:
  k = 0
  for i in nums:
    if i > 0:
      k = k+1
  return k
```

```python
print(count_positive(spisok))
```

```python
def sum_even(nums: list[int]) -> int:
  s = 0
  for i in nums:
    if i%2 == 0:
      s = s + i
  return s
```

```python
print(sum_even(spisok))
```

```python
def first_divisible(nums: list[int], k: int) -> int | None:
  r = None
  for i in nums:
    if i%k == 0:
      r = i
      break
  return r
```

```python
print(first_divisible(spisok,3))
```

```python
#задание 4
class Chocolate:
  def __init__(self, name: str, price: float) -> None:
    if price <= 0:
      raise ValueError
    self.name = name
    self.price = price
  def discount(self, percent: int) -> float:
     # предполагается, что проценты вводятся в формате от 0 до 100 числом
     return self.price * (100 - percent)/100

```

```python
chocolate = Chocolate('Аленка',160)
chocolate.discount(25)
```
