- Посчитайте процент выживаемости у каждого класса пассажиров (Pclass) - **2 балла**

```python
#можно просуммировать, так как 'Survived' принимает значения 1/0
surv_pclass = df.groupby('Pclass').sum('Survived')/df.groupby('Pclass').count() * 100
surv_pclass['Survived']
```

- Выведите самое популярное мужское и самое популярное женское имя на корабле - **3 балла**

```python
#Внимательно изучив данные, увидим, что в признаке Name на самом деле лежит полное имя, поэтому в начале создадим признак Name
df['Full Name'] = df['Name'].copy()
df['Name'] = df['Full Name'].apply(lambda x: x.split(',')[0])
#оставляю только определенный пол и считаю кол-во имен, сортирую по неубыванию
df[df['Sex']=='male']['Name'].value_counts(ascending=False)[:1]
```

```python
df[df['Sex']=='female']['Name'].value_counts(ascending=False)[:1]
```

- Выведите самое популярное мужское и самое популярное женское имя на корабле в каждом классе - **3 балла**

```python
#повторим операцию, но для каждого класса с использованием группировки
name_sex_pclass = df.groupby(['Pclass', 'Sex', 'Name']).count()
```

```python
name_sex_pclass.loc[name_sex_pclass.groupby(['Pclass', 'Sex'])['PassengerId'].idxmax()]['PassengerId']
```

- Выведите количества n-местных кабин (в которых было 2, 3, 4, ... человека) - **2 балла**

- Выведите количество пассажиров, у которых нет родственников на борту - **2 балла**

sibsp: The dataset defines family relations in this way...

parch: The dataset defines family relations in this way...
Parent = mother, father

Some children travelled only with a nanny, therefore parch=0 for them.

Поэтому ищем пассажиров, у которых нет ни сиблингов ни парентов
