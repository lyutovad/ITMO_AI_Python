```python
import pandas as pd

#загрузка файла
df = pd.read_csv('train.csv')
```

**Выведите основную информацию о датасете: информацию о типах данных, число пропусков, средние значения и т.д. - 2 балла**

Посчитайте процент выживаемости у каждого класса пассажиров (Pclass) - 2 балла

```python
survival_by_class = df.groupby('Pclass')['Survived'].mean() * 100 #здесь мы берем конкретную колонку конкретным признаком и находим среднее
print(survival_by_class)
```

**Выведите самое популярное мужское и самое популярное женское имя на корабле - 3 балла**

```python
def extract_first_name(name):
    #чистим данные столбца от фамилий и mrs.ms.и т.д.
    first_name = name.split(',')[1].split('.')[1].strip()
    #если есть имя в скобках,то берем его
    if '(' in first_name:
        first_name = first_name.split('(')[1].split(' ')[0].replace(')', '')
    else:
        first_name = first_name.split(' ')[0]
    return first_name

#создаем временный столбец с "чистыми" именами
df['FirstName'] = df['Name'].apply(extract_first_name)

#популярные имена в целом, туту мы по столбцу пола берем конкретные мужские и женские 
top_male_name = df[df['Sex'] == 'male']['FirstName'].mode()[0]
top_female_name = df[df['Sex'] == 'female']['FirstName'].mode()[0]

print(f"самое популярное мужское имя: {top_male_name}")
print(f"самое популярное женское имя: {top_female_name}")
```

**Выведите самое популярное мужское и самое популярное женское имя на корабле в каждом классе - 3 балла**

```python
#популярные имена в разрезе каждого класса, 
for p_class in sorted(df['Pclass'].unique()):
    m_name = df[(df['Pclass'] == p_class) & (df['Sex'] == 'male')]['FirstName'].mode()[0]
    f_name = df[(df['Pclass'] == p_class) & (df['Sex'] == 'female')]['FirstName'].mode()[0]
    print(f"класс {p_class}: мужское — {m_name}, женское — {f_name}")
```

**for p_class in sorted(df['Pclass'].unique()):**
находит все уникальные значения в колонке Pclass, сортирует их по порядку и начинает цикл

внутри цикла создаются две выборки:

**df[(df['Pclass'] == p_class) & (df['Sex'] == 'male')]** - только мужчины выбранного класса.

**df[(df['Pclass'] == p_class) & (df['Sex'] == 'female')]** - только женщины выбранного класса.

поиск моды (.mode()[0]):
метод .mode() вычисляет самое часто встречающееся значение в колонке FirstName. Так как мода может вернуть несколько значений, то берется самое первое из них - индекс [0].

**Выведите количество пассажиров, у которых нет родственников на борту - **2 балла** **

```python
#считаем n-местные кабины
#сначала считаем, сколько раз встречается номер каждой кабины,dropna убирает пустые значения
#мы смотрим, сколько раз номер каждой конкретной кабины (например, «C23») встречается в списке. 
#если «C23» встречается 3 раза, значит, в ней ехало 3 пассажира.

occupancy = df['Cabin'].dropna().value_counts()

#тут затем считаем частоту этих повторений
#если у нас есть пять разных кабин, в каждой из которых сидело по 2 человека, 
# то для n=2 (двухместные) результат будет count=5

n_place_cabins = occupancy.value_counts().sort_index()

print("Количество кабин с разным числом пассажиров:")
for n, count in n_place_cabins.items():
    print(f"{n}-местных кабин: {count}")

#пассажиры без родственников
#SibSp - братья/сестры/супруги, Parch - родители/дети. Если оба 0 — человек один.
no_fam = df[(df['SibSp'] == 0) & (df['Parch'] == 0)]
print(f"\nКоличество пассажиров без родственников на борту: {len(no_fam)}")
```

(df['SibSp'] == 0) & (df['Parch'] == 0)

Если оба этих значения равны 0, значит, человек путешествовал без близких родственников, потом код фильтрует таблицу по этим условиям и выводит общее количество таких строк через len()
