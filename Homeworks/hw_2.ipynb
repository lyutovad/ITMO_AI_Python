**4 Пункт**

Признаюсь честно, для выяснения, что из записи является именем, обратился к ИИ. При выполнении задания предполагаю, что первое слово после точки в обращении - имя. Для женщин с двойным именем - первое слово после скобок.

Создадим функцию для извлечения имени.

```python
def get_name (full_name):
  name_p1 = full_name.split(',')[1].strip()
  if '(' in name_p1:
     in_brackets = name_p1.split('(')[1].split(')')[0]
     name = in_brackets.split()[0]
  else:
    after_dot = after_title = name_p1.split('.')[1].strip()
    name = after_title.split()[0]
  return name
```

теперь с apply применим ф-ю к каждому пассажиру

```python
df['Only_name'] = df['Name'].apply(get_name)
df['Only_name']
```

Теперь найдём самое популярное имя для каждого пола

```python
male_n = df[df['Sex'] == 'male']['Only_name'].value_counts()
popular_m = male_n.index[0]
popular_m
#мужчины
```

```python
female_n = df[df['Sex'] == 'female']['Only_name'].value_counts()
popular_fem = female_n.index[0]
popular_fem
```

**Пункт 8**

В таблице не указано, сколько мест в кабинах, поэтому потребуется вычислить сколько пассажиров жили вместе, сгруппировать по кабинам, а затем посмотреть размер каждой кабины

```python

pass_w_cab = df[df['Cabin'].notna()] ## уберем пассажиров без кабины
```

```python
## сгруппируем пассажиров по кабинам, посмотрим размер каждой кабины
cab_size =  pass_w_cab.groupby('Cabin').size()
cab_size
```

```python
## посчитаем сколько кабин какого типа
cab_count = cab_size.value_counts().sort_index()
cab_count
```
