# Работа со строками и исключениями
Создайте функцию:

```python
def parse_int(s: str) -> int | None:
    ...
```
## Требования:

1. Функция должна пытаться преобразовать строку `s` в `int`
2. Перед преобразованием необходимо удалить лишние пробелы (`strip()`)
3. Если преобразование невозможно — функция не должна падать, а должна возвращать `None`
4. Использовать `try/except ValueError`

## Продемонстрировать работу функции:

- `"12"`
- `" -7 "`
- `"12a"`
- `""`

````python
def parse_int(s: str) -> int | None:
  s = s.strip()
  try:
    return int(s)
  except ValueError:
    return None

doublecheck = ["12", " -7 ", "12a", ""]

for example in doublecheck:
    result = parse_int(example)
    print(f"Проверка: '{example}' -> Вывод функции: {result}")
````

# Циклы и работа со списками

Создайте список чисел (20–30 элементов).

Реализуйте функции:

```python
def count_positive(nums: list[int]) -> int:
    ...

def sum_even(nums: list[int]) -> int:
    ...

def first_divisible(nums: list[int], k: int) -> int | None:
    ...
```

## Требования

### 1. count_positive  
Возвращает количество чисел больше 0.

### 2. sum_even  
Возвращает сумму всех чётных чисел.

### 3. first_divisible  
Возвращает первое число, которое делится на `k`.  
Если таких чисел нет — вернуть `None`.

Использовать:
- цикл `for`
- оператор `break`

````python
numbers = [5, -3, 12, 0, 7, -18, -10, 4, 9, 21,
           -6, 14, 3, -2, 8, 11, -15, 6, 2, -1,
           19, 24, -8, 13, 16]


def count_positive(nums: list[int]) -> int:
    count = 0
    for num in nums:
        if num > 0:
            count += 1
    return count


def sum_even(nums: list[int]) -> int:
    total = 0
    for num in nums:
        if num % 2 == 0:
            total += num
    return total


def first_divisible(nums: list[int], k: int) -> int | None:
    result = None
    for num in nums:
        if k != 0 and num % k == 0:
            result = num
            break
    return result


print("Количество положительных:", count_positive(numbers))
print("Сумма чётных:", sum_even(numbers))
print("Первое число, делящееся на 3:", first_divisible(numbers, 3))
print("Первое число, делящееся на 50:", first_divisible(numbers, 50))
````

# Lambda и сортировки

Дан список словарей:

```python
items = [
  {"name": "A", "price": 120, "rating": 4.7},
  {"name": "B", "price": 90, "rating": 4.2},
  {"name": "C", "price": 150, "rating": 4.9},
]
```

Необходимо:

1. Отсортировать список по `price` (по возрастанию).
2. Отсортировать список по `rating` (по убыванию).

Использовать `sorted(..., key=lambda ...)`.

````python
items = [
    {"name": "A", "price": 120, "rating": 4.7},
    {"name": "B", "price": 90, "rating": 4.2},
    {"name": "C", "price": 150, "rating": 4.9},
]

sorted_by_price = sorted(items, key=lambda x: x["price"])
sorted_by_rating = sorted(items, key=lambda x: x["rating"], reverse=True)

print("Сортировка по цене (возрастание):")
for item in sorted_by_price:
    print(item)

print("\nСортировка по рейтингу (убывание):")
for item in sorted_by_rating:
    print(item)
````

# Мини-ООП

Создайте класс:

```python
class Chocolate:
    ...
```

## Требования

1. Поля:
   - `name: str`
   - `price: float`
2. Если `price <= 0` — выбросить `ValueError`
3. Метод:

```python
def discount(self, percent: int) -> float:
    ...
```

Метод должен возвращать новую цену со скидкой (не обязательно менять поле).

## Продемонстрировать:

- создание минимум двух объектов
- вызов метода `discount`

````python
class Chocolate:
    def __init__(self, name: str, price: float):
        if price <= 0:
            raise ValueError("Цена не может быть такой. Напиши положительное число")
        self.name = name
        self.price = price

    def discount(self, percent: int) -> float:
        return self.price * (1 - percent / 100)


chocolate1 = Chocolate("Альпенгольд", 150)
chocolate2 = Chocolate("Милка", 200)

print(f"{chocolate1.name} со скидкой 10%:", chocolate1.discount(10))
print(f"{chocolate2.name} со скидкой 25%:", chocolate2.discount(25))
````
