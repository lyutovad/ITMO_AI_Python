# Часть 1 — Работа со строками и исключениями

Функция `parse_int` пытается преобразовать строку в целое число.

Шаги:
1. Удаляем пробелы с помощью `strip()`
2. Пытаемся преобразовать строку в `int`
3. Если возникает `ValueError`, возвращаем `None`

```python
def parse_int(s: str) -> int | None:
    try:
        s = s.strip()
        return int(s)
    except ValueError:
        return None
```

### Демонстрация работы функции

```python
tests = ["12", " -7 ", "12a", ""]

for t in tests:
    print(f"{t!r} -> {parse_int(t)}")
```

# Часть 2 — Циклы и работа со списками

Создадим список чисел и реализуем три функции:

- `count_positive` — количество положительных чисел
- `sum_even` — сумма чётных чисел
- `first_divisible` — первое число, делящееся на `k`

```python
nums = [5, -3, 12, 8, 0, -7, 14, 22, -1, 9, 6, -10, 3, 4, 17, -5, 18, 20, -2, 11, 7, 16, -8, 2]
```

```python
def count_positive(nums: list[int]) -> int:
    count = 0
    for n in nums:
        if n > 0:
            count += 1
    return count
```

```python
def sum_even(nums: list[int]) -> int:
    total = 0
    for n in nums:
        if n % 2 == 0:
            total += n
    return total
```

```python
def first_divisible(nums: list[int], k: int) -> int | None:
    for n in nums:
        if n % k == 0:
            return n
    return None
```

### Проверка функций

```python
print("Количество положительных:", count_positive(nums))
print("Сумма чётных:", sum_even(nums))
print("Первое число делящееся на 7:", first_divisible(nums, 7))
print("Первое число делящееся на 13:", first_divisible(nums, 13))
```

# Часть 3 — Lambda и сортировка

```python
items = [
    {"name": "A", "price": 120, "rating": 4.7},
    {"name": "B", "price": 90, "rating": 4.2},
    {"name": "C", "price": 150, "rating": 4.9},
]
```

### Сортировка по цене (возрастание)

```python
sorted_by_price = sorted(items, key=lambda x: x["price"])
print(sorted_by_price)
```

### Сортировка по рейтингу (убывание)

```python
sorted_by_rating = sorted(items, key=lambda x: x["rating"], reverse=True)
print(sorted_by_rating)
```

# Часть 4 — Мини ООП

Создадим класс `Chocolate`.

Поля:
- `name`
- `price`

Если цена меньше или равна нулю — выбрасывается `ValueError`.

Метод `discount(percent)` возвращает цену со скидкой.

```python
class Chocolate:

    def __init__(self, name: str, price: float):
        if price <= 0:
            raise ValueError("Price must be greater than 0")

        self.name = name
        self.price = price

    def discount(self, percent: int) -> float:
        return self.price * (1 - percent / 100)
```

### Демонстрация работы класса

```python
ch1 = Chocolate("Milka", 3.5)
ch2 = Chocolate("Lindt", 5.0)

print(ch1.name, "price with 10% discount:", ch1.discount(10))
print(ch2.name, "price with 20% discount:", ch2.discount(20))
```
