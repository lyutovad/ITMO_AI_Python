#Задание 1

```python
def parse_int(s: str) -> int | None:
  try:
    return int(s.strip())
  except ValueError:
    return None
```

```python
a=["12"," -7 ","12a",""]
for i in a:
  c = parse_int(i)
  print(f'"{i}" -> {c}')
```

#Задание 2

```python
nums = [12, -5, 7, 0, 18, -2, 9, 4, -11, 25, 6, 14, -8, 3, 20, -1, 16, 5, -7, 10]
def count_positive(nums: list[int]) -> int:
    count = 0
    for num in nums:
        if num > 0:
            count += 1
    return count
def sum_even(nums: list[int]) -> int:
    total = 0
    for num in nums:
        if num % 2 == 0:
            total += num
    return total
def first_divisible(nums: list[int], k: int) -> int | None:
    for num in nums:
        if num % k == 0:
            return num
            break
    return None
print("Список чисел:", nums)
print("Количество положительных чисел:", count_positive(nums))
print("Сумма чётных чисел:", sum_even(nums))
print("Первое число, делящееся на 5:", first_divisible(nums, 5))
print("Первое число, делящееся на 13:", first_divisible(nums, 13))
```

#Задание 3

```python
items = [
    {"name": "A", "price": 120, "rating": 4.7},
    {"name": "B", "price": 90, "rating": 4.2},
    {"name": "C", "price": 150, "rating": 4.9},
]

sorted_by_price = sorted(items, key=lambda i: i["price"])

sorted_by_rating = sorted(items, key=lambda i: i["rating"], reverse=True)

print("По price:", sorted_by_price)
print("По rating:", sorted_by_rating)
```

#Задание 4

```python
class Chocolate:
    def __init__(self, name: str, price: float):
        if price <= 0:
            raise ValueError("Цена должна быть больше 0")
        self.name = name
        self.price = price

    def discount(self, percent: int) -> float:
        return self.price - (self.price * percent / 100)


a = Chocolate("Alpen Gold", 100.0)
b = Chocolate("Nestle", 130.0)

print(a.discount(20))
print(b.discount(15))
```
