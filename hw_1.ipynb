```python
def parse_int(s: str) -> int | None:
    try:
        return int(s.strip())
    except ValueError:
        return None
```

Задание 2. Циклы и работа со списками

```python
spisok = [12228, -5, 7, 1, 132138, 2313, -9, 4442, 16, -666, 12, 30, 5, -8, 21, 14, 7, -1, 6, 19, 8, -4, -223232, -424242, 424242]

```

Функция один

```python
def count_positive(spisok: list[int]) -> int:
    cou = 0
    for n in spisok:
        if n > 0:
            cou += 1
    return cou
```

Функция 2

```python
def sum_even(spisok: list[int]) -> int:
    total = 0
    for n in spisok:
        if n % 2 == 0:
            total += n
    return total
```

Функция Three

```python
def first_divisible(spisok: list[int], k: int) -> int | None:
    for n in spisok:
        if n % k == 0:
            return n
    return None
```

Проверка функций............

```python
print("Вот сколько чисел больше 0:", count_positive(spisok))
print("Вот сумма четных чисел:", sum_even(spisok))
print("Первое число делящееся на 152:", first_divisible(spisok, 152))
print("Первое число делящееся на 2:", first_divisible(spisok, 2))
print("Первое число делящееся на 1380:", first_divisible(spisok, 1380))
print("Первое число делящееся на 7:", first_divisible(spisok, 7))
```

Задание 3. Lambda и сортировки

```python
items = [
  {"name": "A", "price": 120, "rating": 4.7},
  {"name": "B", "price": 90, "rating": 4.2},
  {"name": "C", "price": 150, "rating": 4.9},
]
#price по возрастанию
by_price = [item["name"] for item in sorted(items, key=lambda x: x["price"])]
#rating по убываню
by_rating = [item["name"] for item in sorted(items, key=lambda x: x["rating"], reverse=True)]
print("первое:", by_price)
print("второе:", by_rating)
```

Задание 4. мини ООП

создание класса

```python
class Chocolate:
    def __init__(self, name: str, price: float):
        if price <= 0:
            raise ValueError("ЦЕНА МЕНЬШЕ (ИЛИ РАВНА) 0")
        self.name = name
        self.price = price

    def discount(self, percent: int) -> float:
        if percent > 100:
            raise ValueError("СКИДКУ УМЕНЬШИТЬ!!!!")
        return self.price * (1 - percent / 100)
```

тест
