# ЧАСТЬ 1. Работа со строками и исключениями (4 балла)

Создайте функцию:

```python
def parse_int(s: str) -> int | None:
    ...
```

## Требования:

1. Функция должна пытаться преобразовать строку `s` в `int`.
2. Перед преобразованием необходимо удалить лишние пробелы (`strip()`).
3. Если преобразование невозможно — функция не должна падать, а должна возвращать `None`.
4. Использовать `try/except ValueError`.

## Продемонстрировать работу функции:

- `"12"`
- `" -7 "`
- `"12a"`
- `""`

````python
def parse_int(s: str) -> int | None:
 try:
   return int(s.strip())
 except ValueError:
   return None


print(parse_int)
examples=["12", " -7", "12a", ""]
for examples in examples:
    result = parse_int(examples)
    print(f"parse_int({examples}) -> {result}")
    print()
````

# ЧАСТЬ 2. Циклы и работа со списками (6 баллов)

Реализуйте функции:

```python
def count_positive(nums: list[int]) -> int:
    ...

def sum_even(nums: list[int]) -> int:
    ...

def first_divisible(nums: list[int], k: int) -> int | None:
    ...
```

## Требования

### 1. count_positive  
Возвращает количество чисел больше 0.

### 2. sum_even  
Возвращает сумму всех чётных чисел.

### 3. first_divisible  
Возвращает первое число, которое делится на `k`.  
Если таких чисел нет — вернуть `None`.

Использовать:
- цикл `for`
- оператор `break`

### Оценивание

- `count_positive` — **2 балла**
- `sum_even` — **2 балла**
- `first_divisible` (с `break` и корректным `None`) — **2 балла**

````python
def count_positive(nums: list[int]) -> int:
    count = 0
    for num in nums:
        if num > 0:
            count += 1
    return count

def sum_even(nums: list[int]) -> int:
    total = 0
    for num in nums:
        if num % 2 == 0:
            total += num
    return total

def first_divisible(nums: list[int], k: int) -> int | None:
    for num in nums:
        if num != 0 and num % k == 0:
            return num
    return None


print()
numbers = [-15, 8, -3, 12, 0, 7, -9, 24, 5, -2,
           18, -11, 30, 4, -6, 9, -1, 16, -4, 20,
           -8, 25, -12, 14, -5]

print(f"Список чисел: {numbers}")
print(f"Количество положительных: {count_positive(numbers)}")

positive_count = count_positive(numbers)
print(f"Проверим: положительные числа - {[n for n in numbers if n > 0]}")
print(f"Количество: {positive_count}")

print(f"\nСумма чётных чисел: {sum_even(numbers)}")

even_numbers = [n for n in numbers if n % 2 == 0]
even_sum = sum(even_numbers)
print(f"Чётные числа: {even_numbers}")
print(f"Их сумма: {even_sum}")
````

# ЧАСТЬ 3. Lambda и сортировки (2 балла)

Дан список словарей:

```python
items = [
  {"name": "A", "price": 120, "rating": 4.7},
  {"name": "B", "price": 90, "rating": 4.2},
  {"name": "C", "price": 150, "rating": 4.9},
]
```

Необходимо:

1. Отсортировать список по `price` (по возрастанию).
2. Отсортировать список по `rating` (по убыванию).

Использовать `sorted(..., key=lambda ...)`.

### Оценивание

- корректная сортировка по `price` — **2 балл**
- корректная сортировка по `rating` — **2 балл**

````python
print()
items = [
    {"name": "A", "price": 120, "rating": 4.7},
    {"name": "B", "price": 90, "rating": 4.2},
    {"name": "C", "price": 150, "rating": 4.9},
]

sorted_by_price = sorted(items, key=lambda x: x["price"])
print()
for item in sorted_by_price:
    print(f"  {item['name']}: price={item['price']}, rating={item['rating']}")

sorted_by_rating = sorted(items, key=lambda x: x["rating"], reverse=True)
print()
for item in sorted_by_rating:
    print(f"  {item['name']}: price={item['price']}, rating={item['rating']}")
print()
````
