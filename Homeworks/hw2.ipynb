## Процент выживаемости по классу пассажиров

Посчитаем процент выживших пассажиров в каждом классе (`Pclass`).

```python
survival_rate = df.groupby("Pclass")["Survived"].mean() * 100
survival_rate
```

Результат показывает, что вероятность выживания зависит от класса пассажира.
Как правило, пассажиры первого класса имели более высокий процент выживания.

## Самые популярные мужские и женские имена

Из столбца `Name` извлечем имя пассажира.

```python
df["FirstName"] = df["Name"].str.extract(r",\s*(?:Mr|Mrs|Miss|Master|Dr)\.\s*([A-Za-z]+)")
df[["Name","FirstName"]].head()
```

```python
male_name = df[df["Sex"]=="male"]["FirstName"].value_counts().idxmax()
female_name = df[df["Sex"]=="female"]["FirstName"].value_counts().idxmax()

male_name, female_name
```

Получили наиболее часто встречающиеся мужские и женские имена среди пассажиров.

## Самые популярные имена по полу и классу

```python
for pclass in sorted(df["Pclass"].unique()):

    male = df[(df["Sex"]=="male") & (df["Pclass"]==pclass)]["FirstName"].value_counts().idxmax()

    female = df[(df["Sex"]=="female") & (df["Pclass"]==pclass)]["FirstName"].value_counts().idxmax()

    print(f"Pclass {pclass}:")
    print(" most popular male name:", male)
    print(" most popular female name:", female)
    print()
```

## Количество n-местных кают

```python
cabins = df["Cabin"].dropna()

cabin_counts = cabins.str.split().apply(len).value_counts().sort_index()

cabin_counts
```

Результат показывает количество кают, в которых размещалось 2, 3, 4 и более пассажиров.
